```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
total_cities = len(cities)
len(cities)
```

```python
#Create empty lists which will be appended to 
date_list = []
city_name_list = []
country_code_list = []
lat_list = []
lng_list = []
temp_list = []
humidity_list =[]
cloudy_list = []
wind_speed_list = []

#Initialize request number
request_number = 1

print('Beginning Data Retreival')

#Loop through cities generated
for city in cities:
    
    #Make use of try/except statement so that program doesnt end when a city is not found
    try:
        
        #API call
        params = {'q': city,
              'units': 'metric',
              'appid': api_key
                 }
        url = "http://api.openweathermap.org/data/2.5/weather?"
        response = requests.get(url,params).json()
        
        print(f'Processing Record {request_number} out of {total_cities} | {city}')
        
        #Append values of interest to their corresponding list
        date = response["dt"]
        date_list.append(date)
        
        city_name = city
        city_name_list.append(city_name)
        
        country_code = response["sys"]["country"]
        country_code_list.append(country_code)
        
        lat = response['coord']['lat']
        lat_list.append(lat)
        
        lng = response['coord']['lon']
        lng_list.append(lng)
        
        temp = response['main']['temp_max']
        temp_list.append(temp)
        
        humidity = response['main']['humidity']
        humidity_list.append(humidity)
        
        cloudy = response["clouds"]["all"]
        cloudy_list.append(cloudy)
        
        wind_speed = response['wind']['speed']
        wind_speed_list.append(wind_speed)
              
    except KeyError:
        print('City not found. Skipping ...')
    
    #Keeping track of request number (add one after every iteration)
    request_number = request_number + 1
```

```python
#Creating a dataframe
weather_data = {'City': city_name_list,
               'Cloudiness': cloudy_list,
               'Country': country_code_list,
               'Date': date_list,
               'Humidity': humidity_list,
               'Lat': lat_list,
               'Lng': lng_list,
               'Max Temp': temp_list,
               'Wind Speed': wind_speed_list
               }

weather_df = pd.DataFrame(weather_data)
#weather_df.to_csv(output_data_file)
weather_df.head()
```
